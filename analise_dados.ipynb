```python
media_salario = dados['SALARIO'].mean()
desvio_salario = dados['SALARIO'].std()
limite_superior = media_salario + (3 * desvio_salario)
print(limite_superior)
```

```python
# média de cada faixa dos outliers
media_40k = dados[(dados['FAIXA SALARIAL'] == 'Acima de R$ 40.001/mês') & (dados['SALARIO'] <= limite_superior)]['SALARIO'].mean()
media_30_40k = dados[(dados['FAIXA SALARIAL'] == 'de R$ 30.001/mês a R$ 40.000/mês') & (dados['SALARIO'] <= limite_superior)]['SALARIO'].mean()
```

```python
# substituindo os outliers pela média da faixa
dados.loc[(dados['SALARIO'] > limite_superior) & (dados['FAIXA SALARIAL'] == 'Acima de R$ 40.001/mês'), 'SALARIO'] = media_40k
dados.loc[(dados['SALARIO'] > limite_superior) & (dados['FAIXA SALARIAL'] == 'de R$ 30.001/mês a R$ 40.000/mês'), 'SALARIO'] = media_30_40k
```

```python
# confirmando que não tem mais outliers
dados[dados['SALARIO'] > limite_superior]['SALARIO'].count()
```

```python
def determinar_geracao(idade):
    if idade <= 26:
        return 'Geração Z'
    elif idade <= 41:
        return 'Millennial'
    else:
        return 'Geração X'

dados['GERACAO'] = dados['IDADE'].apply(determinar_geracao)
dados['GERACAO'].value_counts()
```

## Correlação

```python
dados['IDADE'].corr(dados['SALARIO'])
```

```python
from scipy.stats import chi2_contingency

def cramer_coeficiente(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    return np.sqrt(phi2 / min(r-1, k-1))

cramer_coeficiente(dados['COR/RACA/ETNIA'], dados['NIVEL DE ENSINO'])
```
